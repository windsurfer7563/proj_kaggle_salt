```python
import pandas as pd
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
f1 = np.load('../classification/logs/fold1/dataset.predictions.infer.logits.npy')
f2 = np.load('../classification/logs/fold2/dataset.predictions.infer.logits.npy')
f3 = np.load('../classification/logs/fold3/dataset.predictions.infer.logits.npy')
f4 = np.load('../classification/logs/fold4/dataset.predictions.infer.logits.npy')
f5 = np.load('../classification/logs/fold5/dataset.predictions.infer.logits.npy')
```

```python
prediction= (sigmoid(f1)+sigmoid(f2)+sigmoid(f3)+sigmoid(f4)+sigmoid(f5))/5
```

```python
pd_sub0 = pd.read_csv('../data/submission.csv')
```
